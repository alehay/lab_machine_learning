```python
y = data.JP_Sales
y
```

```python
x = data.drop(columns = ['Name', 'Developer', 'JP_Sales'])
x.info()
```

```python
x = x.fillna(-100)
```

```python
x
```

```python
x.info()
```

```python
value_counts = x['Platform'].value_counts()
value_counts
```

```python
low_frequency_categories = value_counts[value_counts < 100].index
low_frequency_categories
```

```python
x['Platform'] = x['Platform'].replace(low_frequency_categories, 'Other')
value_counts = x['Platform'].value_counts()
value_counts
```

```python
value_counts = x['Publisher'].value_counts()
value_counts
```

```python
low_frequency_categories = value_counts[value_counts < 150].index
low_frequency_categories
```

```python
x['Publisher'] = x['Publisher'].replace(low_frequency_categories, 'Other')
value_counts = x['Publisher'].value_counts()
value_counts
```

```python
value_counts = x['Genre'].value_counts()
value_counts
```

```python
value_counts = x['Rating'].value_counts()
value_counts
```

```python
low_frequency_categories = value_counts[value_counts < 150].index
low_frequency_categories
```

```python
x['Rating'] = x['Rating'].replace(low_frequency_categories, 'Other')
value_counts = x['Rating'].value_counts()
value_counts
```

```python
x = pd.get_dummies(x, drop_first = True)
x
```

```python
x.info()
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x, y)
y_pred = lr.predict(x)
```

```python
from sklearn.metrics import mean_absolute_error, mean_absolute_percentage_error
print('Mean Absolute Error = ', mean_absolute_error(y, y_pred))
print('Mean Absolute Percentage Error = ', mean_absolute_percentage_error(y, y_pred))
```

```python
from catboost import CatBoostRegressor, Pool
data = pd.read_csv('Video_Games.csv')
data['Genre'] = data['Genre'].fillna(-100)
data['Publisher'] = data['Publisher'].fillna(-100)
data['Rating'] = data['Rating'].fillna(-100)
x = data.drop(columns = ['Name', 'Developer', 'JP_Sales'])
print(x.info())
train_data = Pool(x, y, cat_features = ['Platform', 'Genre', 'Publisher', 'Rating'])
cbr = CatBoostRegressor()
cbr.fit(train_data)
```

```python
y_pred = cbr.predict(train_data)
print('Mean Absolute Error = ', mean_absolute_error(y, y_pred))
print('Mean Absolute Percentage Error = ', mean_absolute_percentage_error(y, y_pred))
```
