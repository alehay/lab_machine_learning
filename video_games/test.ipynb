```python
import pandas as pd
import numpy as np
from sklearn.metrics import mean_absolute_error

train = pd.read_csv("Video_Games.csv")

if "Global_Sales" in train.columns:
    jp_hat = train["Global_Sales"] - train["NA_Sales"] - train["EU_Sales"] - train["Other_Sales"]
    jp_hat = np.clip(jp_hat, 0, None)
    print("MAE(leak baseline) =", mean_absolute_error(train["JP_Sales"], jp_hat))
else:
    print("Global_Sales нет — формула неприменима напрямую.")
```
